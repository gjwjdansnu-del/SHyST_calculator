```python
import matplotlib.pyplot as plt
import numpy as np
import os

# 파란 점 위치
x_positions_blue = [227.5, 297.5, 367.5, 437.5]
angles_blue = [-90, 90]

# 붉은 점 위치
x_positions_red = list(range(210, 456, 35))  # 210, 245, ..., 455
angles_red = [-90, 90]

# 파란 점 좌표 생성
blue_x = []
blue_y = []
for angle in angles_blue:
    for x in x_positions_blue:
        blue_x.append(angle)
        blue_y.append(x)

# 붉은 점 좌표 생성
red_x = []
red_y = []
for angle in angles_red:
    for x in x_positions_red:
        red_x.append(angle)
        red_y.append(x)

# 플롯
fig, ax = plt.subplots(figsize=(5, 5))

# 배경 음영
ax.axvspan(-180, 0, color='red', alpha=0.1)
ax.axvspan(0, 180, color='blue', alpha=0.1)

# 점 플로팅 (크기 지정)
ax.scatter(blue_x, blue_y, color='blue', label='Pressure tap', s=80)
ax.scatter(red_x, red_y, color='red', label='Temperature tap', s=40)

# 축 설정
ax.set_xlabel("Azimuthal Angle [°]")
ax.set_ylabel("x [mm]")
ax.set_xlim(-180, 180)
ax.set_ylim(200, 465)
ax.set_xticks(np.linspace(-180, 180, 5))
ax.set_yticks(np.linspace(210, 455, 15))
ax.grid(True)

# 범례
ax.legend(loc='upper center', fontsize=10)

# 저장 및 출력
plt.tight_layout()
save_path = os.path.join(os.getcwd(), "Tap_Location_Map.pdf")
plt.savefig(save_path, dpi=600, bbox_inches='tight')
plt.show()

print(f"✅ 고화질 PDF 저장 완료: {save_path}")
```
